```python
import random  # 계좌번호 랜덤하게 생성하는 randint 함수 사용을 위해 random import함

# Q1. 클래스 생성
class Account:
    # 클래스 속성
    bank_name = "SC은행"
    account_count = 0  # 생성된 계좌 객체의 수를 저장하기 위한 클래스 변수

    def __init__(self, account_holder, initial_balance):
        self.account_holder = account_holder
        self.balance = initial_balance  # 계좌의 초기 잔액
        # 랜덤한 계좌 번호 생성 (111-11-111111 형식)
        self.account_number = f"{random.randint(100, 999)}-{random.randint(10, 99)}-{random.randint(100000, 999999)}"
        Account.account_count += 1  # Q2. 계좌의 객체의 개수 저장
        self.deposit_count = 0     # Q7. 입금횟수를 확인하기 위해 기초값을 0으로 설정
        # 입출금 기록을 저장할 리스트
        self.deposit_hist = []
        self.withdraw_hist = []

    # Q3. 계좌의 개수 출력 메서드
    def get_account_num(self):
        print(f"생성된 계좌 수: {Account.account_count}")
        print()

    # Q4.입금메서드
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount     #입금 잔액 증가
            self.deposit_hist.append(amount)    # 입금 기록에 추가
            self.deposit_count += 1    #입금 횟수 증가
            # Q7.5회 입금 시 1% 이자 지급
            if self.deposit_count == 5:
                print(f"입금액: {amount:,} 원. 현재 잔액: {self.balance:,} 원.")
                interest = round(self.balance*0.01)
                self.balance += interest
                self.deposit_count = 0  # 5회 이자 지급을 완료하면 다시 카운트 하기 위해 초기화시킴
                print(f"5회 입금 이자금액: {interest:,} 원. 현재 잔액: {self.balance:,} 원.")
            else:
                print(f"입금액: {amount:,} 원. 현재 잔액: {self.balance:,} 원.")
        else:
            print("입금액은 최소 1원 이상이어야 합니다.")

    # Q5. 출금메서드
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            self.withdraw_hist.append(amount)  # 출금 기록에 추가
            print(f"출금액: {amount:,} 원. 현재 잔액: {self.balance:,} 원.")
        else:
            print("출금 가능한 잔액이 부족합니다.")


    # Q6. 계좌 정보를 출력
    def display_info(self):
        balance_comma = "{:,}".format(self.balance)  # 잔액을 천 단위로 구분하여 표시
        print(f"은행명: {self.bank_name}")
        print(f"예금주: {self.account_holder}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔액: {balance_comma}원")
        print()


    # Q10  입금내역 기록
    def deposit_history(self):
        print(f"{self.account_holder}의 입금 내역")
        for amount in self.deposit_hist:
            print(f"+ {amount:,}원")
    # 출금 내역 기록
    def withdraw_history(self):
        """출금 내역을 출력하는 메소드."""
        print(f"{self.account_holder}의 출금 내역")
        for amount in self.withdraw_hist:
            print(f"- {amount:,}원")


# Q8.3개 이상 계좌 인스턴스 생성 및 리스트에 추가
accounts = []
accounts.append(Account("RM", 11150000))
accounts.append(Account("뷔", 5000000))
accounts.append(Account("슈가", 7500000))



# 생성된 계좌 수 출력 (인스턴스 메소드를 사용하여 출력)
accounts[0].get_account_num()


# Q9. 잔액이 1백만 원 이상인 계좌만 정보 출력
for account in accounts:
    if account.balance >= 1000000:
        account.display_info()


# 입출금 내역 리스트
test = Account("지민", 5000000)
print("**지민의 입출금 내역 리스트**")
test.deposit(0)
test.deposit(100000)
test.deposit(200000)
test.deposit(300000)
test.deposit(400000)
test.deposit(500000)
test.withdraw(600000)
test.withdraw(1000000000)


# 입출금 내역 출력
print("\n입금 내역:")
test.deposit_history()
print("\n출금 내역:")
test.withdraw_history()



```
