##3. 데이터 전처리

```python
# (1) 이름의 길이가 10 이상인가?

poketmon["name_count"] = poketmon["Name"].apply(lambda i: len(i))

poketmon["long_name"] = poketmon["name_count"] >= 10
display(poketmon.head())
```

```python
# (2) 이름에 자주 쓰이는 토큰 추출

poketmon["Name_nospace"] = poketmon["Name"].apply(lambda i: i.replace(" ", "")) #띄어쓰기 제거

poketmon["name_isalpha"] = poketmon["Name_nospace"].apply(lambda i: i.isalpha()) #  알파벳으로만 이루어졌는지 확인
display(poketmon.head())

display(poketmon[poketmon["name_isalpha"] == False].shape) # 알파벳이 아닌 다른 문자가 이름에 포함된 것
poketmon[poketmon["name_isalpha"] == False]

poketmon = poketmon.replace(to_replace="Nidoran♀", value="Nidoran X")
poketmon = poketmon.replace(to_replace="Nidoran♂", value="Nidoran Y")
poketmon = poketmon.replace(to_replace="Farfetch'd", value="Farfetchd")
poketmon = poketmon.replace(to_replace="Mr. Mime", value="Mr Mime")
poketmon = poketmon.replace(to_replace="Porygon2", value="Porygon Two")
poketmon = poketmon.replace(to_replace="Ho-oh", value="Ho Oh")
poketmon = poketmon.replace(to_replace="Mime Jr.", value="Mime Jr")
poketmon = poketmon.replace(to_replace="Porygon-Z", value="Porygon Z")
poketmon = poketmon.replace(to_replace="Zygarde50% Forme", value="Zygarde Forme")

display(poketmon.loc[[34, 37, 90, 131, 252, 270, 487, 525, 794]])

poketmon["Name_nospace"] = poketmon["Name"].apply(lambda i: i.replace(" ", ""))
poketmon["name_isalpha"] = poketmon["Name_nospace"].apply(lambda i: i.isalpha())
poketmon[poketmon["name_isalpha"] == False]
```

```python
# 이름을 띄어쓰기 & 대문자 기준으로 분리해 토큰화하기

import re  # 정규식 re 패키지를 사용

name = "CharizardMega Charizard X"

name_split = name.split(" ") # 공백을 기준으로 쪼개

temp = name_split[0]

tokens = re.findall('[A-Z][a-z]*', temp) #대문자로 시작해서 소문자로 끝나는 패턴찾기

def tokenize(name):      # 함수로 만들기
        name_split = name.split(" ")     # 공백 기준으로 이름 분리
        tokens = []
        for part_name in name_split:         # 대문자를 기준으로 각 부분 토큰화
          a = re.findall('[A-Z][a-z]*', part_name)
          tokens.extend(a)

        return np.array(tokens)

name = "CharizardMega Charizard X"  #함수 테스
tokenize(name)

# 전체 데이터 셋에 토큰화 함수 적용하기
all_tokens = list(legendary["Name"].apply(tokenize).values)

token_set = []
for token in all_tokens:
    token_set.extend(token)

print(len(set(token_set)))
print(token_set)


from collections import Counter # 토큰이 사용된 개수 확인
a = [1, 1, 0, 0, 0, 1, 1, 2, 3]
Counter(a)   # 각 요소가 몇 개인지를 담은 튜플 리스트를 반환
Counter(a).most_common()  # 가장 많은 요소와 등장 횟수가 무엇인지를 정렬
most_common = Counter(token_set).most_common(10) # most_common()함수에 인자로 숫자를 넣어주면 원하는 개수만큼만 출력
print(most_common)


# 전설의 포켓몬 이름에 등장하는 토큰이 포켓몬의 이름에 있는지의 여부를 나타내는 컬럼

for token, _ in most_common:
    # poketmon[token] = ... 형식으로 사용하면 뒤에서 warning이 발생합니다
    poketmon[f"{token}"] = poketmon["Name"].str.contains(token)

poketmon.head(10)
```

```python
# (3) Type1 & 2! 범주형 데이터 전처리하기
## 18가지의 모든 Type를 모두 원-핫 인코딩(One-Hot Encoding)합니다.
## 두 가지 속성을 가진 포켓몬은 두 가지 Type에 해당하는 자리에서 1 값을 가지도록 합니다

print(types)

for t in types:  # 원핫인코딩
    poketmon[t] = (poketmon["Type 1"] == t) | (poketmon["Type 2"] == t)

poketmon[[["Type 1", "Type 2"] + types][0]].head()
```

##4. 모델학습 및 평가


### 1. 베이스라인


```python
# 원본데이터 불러오기
print(original_data.shape)
original_data.head()
```

```python
original_data.columns
```

```python
# 컬럼 중 의미 없는 컬럼인 #와 문자열 데이터인 Name, Type 1, Type 2 데이터는 제외하고 사용
features = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']
target = 'Legendary'

X = original_data[features]
print(X.shape)

y = original_data[target]
print(y.shape)
```

```python
# 학습 데이터와 테스트 데이터가 분리

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

```

### 2. 모델학습시키기(의사결정트리)


```python
#의사결정모델

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=25)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# 모델의 성능을 평가(accuracy, Confusion Matrix 사용)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# 모델의 성능을 평가(classification_report)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

> 포켓몬 게임

Q. 데이터셋에는 총 몇 가지의 정보가 담겨있으며, 각각은 무엇을 뜻하나요?    
>  총 11가지의 정보가 들어있음.
1. #: 포켓몬의 ID
2. Name: 포켓몬의 이름
3. Type 1: 모든 포켓몬은 타입을 가지고 있으며, 이것은 공격의 약점/저항력을 결정한다.
4. Type 2: 일부 포켓몬은 이중 타입을 가지고 있어 2개의 타입을 가질 수 있다.
5.Total: 이 항목 이후의 모든 스탯의 합계로, 포켓몬의 강도를 일반적으로 나타낸다.
6. HP: 체력 또는 생명력, 포켓몬이 기절하기 전에 얼마나 많은 피해를 버틸 수 있는지를 나타낸다.
7. Attack: 일반 공격(예: 스크래치, 펀치)에 대한 기본 수정자이다.
