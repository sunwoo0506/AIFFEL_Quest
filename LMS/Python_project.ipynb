```python
# 사각형 넓이 구하기

class Square:
    def __init__(self):
        self.square_type = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square_type == 1:
            print('직사각형 함수는 rect()입니다.')
            result =self.rect()
            print(result)
        elif self.square_type == 2:
            print('평행사변형 함수는 par()입니다.')
            result =self.par()
            print(result)
        elif self.square_type == 3:
            print('사다리꼴 함수는 trape()입니다.')
            result =self.trape()
            print(result)

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        base, height = map(int, input('밑변, 높이를 입력하세요. 예시: 밑변,높이\n>>>').split(','))
        area = base * height
        result = '평행사변형의 넓이는 :' +str(area)
        return result

    def trape(self):
        base, top, height = map(int, input('밑변, 윗변, 높이를 입력하세요. 예시: 밑변,윗변,높이\n>>>').split(','))
        area = (base+ top) * height / 2
        result = '사다리꼴의 넓이는 :' +str(area)
        return result

a = Square()
```

```python
#키오스크(kiosk)만들기

# 전체 메뉴와 가격
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

class Kiosk:
    def __init__(self):
        self.menu = menu
        self.price = price


    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    def menu_select(self):
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트
        total_price = 0

        while True:
            print()
            n = int(input("음료를 번호를 입력하세요 (주문 완료: 0): "))  # 음료 번호 입력

            if n == 0:
                break  # 주문 완료 시 종료

            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n <= len(self.menu):
                self.order_price.append(self.price[n-1])  # 가격 리스트에 추가합니다.
                total_price += self.price[n-1]  # 합계 금액

                # 음료 온도 물어보기
                while True:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요: "))
                    if t == 1:
                        temp = "HOT"
                        break
                    elif t == 2:
                        temp = "ICE"
                        break
                    else:
                        print("1과 2 중 하나를 입력하세요.")

                self.order_menu.append(temp + ' ' + self.menu[n-1])  # 주문 리스트에 추가합니다.
                print('주문 음료', temp, self.menu[n-1], ' : ', self.price[n-1], '원')  # 주문 내역 출력
                print("￣"*20)
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 주문이 완료되었습니다.
        print()
        print("주문이 완료되었습니다.")
        print()
        print(self.order_menu, "\n" , self.order_price)  # 주문 내역 출력
        print("총 합계:", sum(self.order_price), "원")

    def pay(self):
        while True:
            pay_method = int(input("지불하실 방법을 선택하세요 (현금: 1, 카드: 2): "))

            if pay_method == 1:
                print("직원을 호출하겠습니다.")
                break
            elif pay_method == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("다시 결제를 시도해주세요.")

    def table(self):
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        print('합계 금액 :', sum(self.order_price))

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')

a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
